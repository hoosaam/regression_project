```python
import pandas as pd
```

read data

```python
data = pd.read_csv(r"D:\coursess\Ai\code_test\Airline_Delay_Cause.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.drop(['carrier','carrier_name','airport','airport_name'],axis=1,inplace=True)
```

```python
data.info()
```

```python
X=data.drop(['carrier_delay'],axis=1,inplace=False)
X.head()
```

```python
y=data['carrier_delay']
list(y)
```

#### Split the data into training and testing datasets.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,random_state=42,shuffle=True)

#Shuffle  might not be necessary if the data is already randomized
#random_state ensures reproducibility
```

```python
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

### Training model

```python
%%time
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge #Mean Squared Error:0.049500115331507645
from sklearn.linear_model import Lasso # Mean Squared Error: 0.9312071252823221
from xgboost import XGBRegressor   #Mean Squared Error: 809222.0879203691

# model=XGBRegressor(n_estimators=1000, learning_rate=0.01, max_depth=5, random_state=42, n_jobs=-1)

model = Ridge(alpha=900.0)  #Mean Squared Error: 6.745507393611007e-05




# model = make_pipeline(StandardScaler(), LinearRegression(fit_intercept=True, copy_X=True, n_jobs=-1))
```

```python
%%time
model.fit(X_train,y_train)
```

### calculate Erorr

```python
y_pred=model.predict(X_test) # type: ignore
```

```python
list(y_pred.round(4))[:10]
```

```python
list(y_test.round(4))[:10]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
MSE= mean_squared_error(y_test, y_pred,multioutput='uniform_average')
print('Mean Squared Error:', MSE)
#0.1391748407838527 linear regression
#0.15105893713171833 
#0.1268420128395927  lasso regression
#0.049500115331507645 ridge regression
#Mean Squared Error: 0.000808968920728767
#Mean Squared Error: 6.745507393611007e-05

```

```python
mse=r2_score(y_test, y_pred,multioutput='uniform_average')
print('R2 Score:', mse)
```

```python
import numpy as np
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Actual', marker='o')
plt.plot(y_pred, label='Predicted', marker='x')
plt.title('Actual vs Predicted Carrier Delay')
plt.xlabel('Sample Index')
plt.ylabel('Carrier Delay')
plt.legend()
plt.grid(True)
plt.show()
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=30)
plt.title('Distribution of Residuals')
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
